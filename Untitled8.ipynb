```python
import os
import xml.etree.ElementTree as ET

# === Paths ===
xml_dir = "/content/allpartslbld"   # folder with all XML files
output_dir = "/content/DronePartsDataset/train/labels/"
os.makedirs(output_dir, exist_ok=True)

# === Unified class list ===
classes = ["propeller", "frame", "broken drone", "drone camera", "broken frame"]


def safe_float(x):
    try:
        return float(x)
    except:
        return None

bad_files = []

# === Convert ===
for xml_file in os.listdir(xml_dir):
    if not xml_file.endswith(".xml"):
        continue

    xml_path = os.path.join(xml_dir, xml_file)

    try:
        tree = ET.parse(xml_path)
        root = tree.getroot()

        w = int(root.find("size/width").text)
        h = int(root.find("size/height").text)

        # Skip corrupted XMLs
        if w == 0 or h == 0:
            bad_files.append(xml_file)
            print(f"âš ï¸ Skipped (invalid image size): {xml_file}")
            continue

        txt_path = os.path.join(output_dir, xml_file.replace(".xml", ".txt"))
        with open(txt_path, "w") as f:

            for obj in root.findall("object"):
                cls_name = obj.find("name").text.lower().strip()

                if cls_name not in classes:
                    print(f"âš ï¸ Unknown class '{cls_name}' in {xml_file}, skipping.")
                    continue

                cls_id = classes.index(cls_name)

                box = obj.find("bndbox")

                xmin = safe_float(box.find("xmin").text)
                ymin = safe_float(box.find("ymin").text)
                xmax = safe_float(box.find("xmax").text)
                ymax = safe_float(box.find("ymax").text)

                # Skip invalid boxes
                if None in [xmin, ymin, xmax, ymax] or xmax <= xmin or ymax <= ymin:
                    print(f"âš ï¸ Bad box in {xml_file}, skipping object.")
                    continue

                x_center = ((xmin + xmax) / 2) / w
                y_center = ((ymin + ymax) / 2) / h
                box_w = (xmax - xmin) / w
                box_h = (ymax - ymin) / h

                f.write(f"{cls_id} {x_center:.6f} {y_center:.6f} {box_w:.6f} {box_h:.6f}\n")

    except Exception as e:
        bad_files.append(xml_file)
        print(f"âŒ Error reading {xml_file}: {e}")

print("\nâœ… Re-conversion complete with unified class mapping!")
if bad_files:
    print("\nâš ï¸ The following files were skipped due to errors:")
    for bf in bad_files:
        print("   -", bf)
```

```python
import os
import xml.etree.ElementTree as ET

xml_dir = "/content/allpartslbld"
output_dir = "/content/DronePartsDataset/train/labels/"
os.makedirs(output_dir, exist_ok=True)

classes = ["propeller", "frame", "broken drone", "drone camera"]

for xml_file in os.listdir(xml_dir):
    if not xml_file.endswith(".xml"):
        continue

    xml_path = os.path.join(xml_dir, xml_file)
    tree = ET.parse(xml_path)
    root = tree.getroot()

    try:
        w = int(root.find("size/width").text)
        h = int(root.find("size/height").text)
    except Exception:
        print(f"âš ï¸ Skipping {xml_file} (missing size info)")
        continue

    if w == 0 or h == 0:
        print(f"âš ï¸ Skipping {xml_file} (width/height = 0)")
        continue

    txt_path = os.path.join(output_dir, xml_file.replace(".xml", ".txt"))
    with open(txt_path, "w") as f:
        for obj in root.findall("object"):
            cls_name = obj.find("name").text
            if cls_name not in classes:
                continue
            cls_id = classes.index(cls_name)

            box = obj.find("bndbox")
            xmin = float(box.find("xmin").text)
            ymin = float(box.find("ymin").text)
            xmax = float(box.find("xmax").text)
            ymax = float(box.find("ymax").text)

            x_center = ((xmin + xmax) / 2) / w
            y_center = ((ymin + ymax) / 2) / h
            box_w = (xmax - xmin) / w
            box_h = (ymax - ymin) / h

            f.write(f"{cls_id} {x_center} {y_center} {box_w} {box_h}\n")

print("âœ… Conversion complete! YOLO labels saved to:", output_dir)
```

```python
import os, random, shutil

# --- paths ---
images_dir = "/content/allparts"           # folder with all images
labels_dir = "/content/DronePartsDataset/train/labels"  # folder with YOLO .txt labels
output_base = "/content/DronePartsDataset"

# --- create folder structure ---
for subset in ["train", "valid"]:
    os.makedirs(f"{output_base}/{subset}/images", exist_ok=True)
    os.makedirs(f"{output_base}/{subset}/labels", exist_ok=True)

# --- split images ---
split_ratio = 0.8
images = [f for f in os.listdir(images_dir)
          if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
random.shuffle(images)
split = int(len(images) * split_ratio)
train, valid = images[:split], images[split:]

# --- safe copy function to prevent SameFileError ---
def safe_copy(src, dst):
    if os.path.abspath(src) != os.path.abspath(dst):  # avoid copying onto itself
        shutil.copy(src, dst)

# --- main copy function ---
def copy_split(files, subset):
    for img in files:
        name = os.path.splitext(img)[0]
        src_img = os.path.join(images_dir, img)
        src_lbl = os.path.join(labels_dir, name + ".txt")
        dst_img = os.path.join(output_base, subset, "images", img)
        dst_lbl = os.path.join(output_base, subset, "labels", name + ".txt")

        # copy image
        if os.path.exists(src_img):
            safe_copy(src_img, dst_img)
        else:
            print(f"âš ï¸ Missing image: {src_img}")

        # copy label (optional)
        if os.path.exists(src_lbl):
            safe_copy(src_lbl, dst_lbl)
        else:
            print(f"âš ï¸ No label found for: {img}")

# --- perform split ---
copy_split(train, "train")
copy_split(valid, "valid")

print("âœ… Dataset split complete.")
```
